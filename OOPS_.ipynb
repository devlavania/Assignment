Q1 What is Object-Oriented Programming (OOP)?  
Ans- A programming paradigm based on objects containing data and behavior.

Q2 What is a class in OOP?  
Ans- A blueprint for creating objects, defining attributes and methods.

Q3 What is an object in OOP?  
Ans- An instance of a class with its own data and behavior.

Q4 What is the difference between abstraction and encapsulation?  
Ans- Abstraction hides complexity; encapsulation hides data using access controls.

Q5 What are dunder methods in Python?  
Ans- Special methods with double underscores (e.g., __init__, __str__) used for operator overloading and object behavior.

Q6 Explain the concept of inheritance in OOP.  
Ans- It allows a class to inherit properties and methods from another class.

Q7 What is polymorphism in OOP?  
Ans- The ability of different classes to respond to the same method call in different ways.

Q8 How is encapsulation achieved in Python?  
Ans- By using private variables (prefixing with _) and property decorators.

Q9 What is a constructor in Python?  
Ans- The __init__ method that initializes object attributes when an object is created.

Q10 What are class and static methods in Python?  
Ans- Class methods use @classmethod and access class state; static methods use @staticmethod and don’t access class or instance state.

Q11 What is method overloading in Python?  
Ans- Python doesn’t support true overloading; it’s simulated using default arguments or *args/**kwargs.

Q12 What is method overriding in OOP?  
Ans- Redefining a parent class method in a child class to change its behavior.

Q13 What is a property decorator in Python?  
Ans- @property allows a method to be accessed like an attribute, enabling controlled access.

Q14 Why is polymorphism important in OOP?  
Ans- It enhances flexibility and code reusability by allowing unified interfaces.

Q15 What is an abstract class in Python?  
Ans- A class with abstract methods defined using the abc module; cannot be instantiated.

Q16 What are the advantages of OOP?  
Ans- Modularity, reusability, scalability, and better code organization.

Q17 What is the difference between a class variable and an instance variable?  
Ans- Class variables are shared across all instances; instance variables are unique to each object.

Q18 What is multiple inheritance in Python?  
Ans- A class inheriting from more than one parent class.

Q19 Explain the purpose of ‘’__str__’ and ‘__repr__’ methods in Python?  
Ans- __str__ returns a readable string; __repr__ returns an unambiguous string for debugging.

Q20 What is the significance of the ‘super()’ function in Python?  
Ans- It allows access to parent class methods and constructors.

Q21 What is the significance of the __del__ method in Python?  
Ans- It’s a destructor called when an object is about to be destroyed.

Q22 What is the difference between @staticmethod and @classmethod in Python?  
Ans- @staticmethod doesn’t take self or cls; @classmethod takes cls and can modify class state.

Q23 How does polymorphism work in Python with inheritance?  
Ans- Child classes override parent methods, and the correct method is called based on the object type.

Q24 What is method chaining in Python OOP?  
Ans- Calling multiple methods on the same object in a single line by returning self.

Q25 What is the purpose of the __call__ method in Python?  
Ans- It makes an object callable like a function.

```python
#2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius**2

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Example usage
# shape = Shape() # This would raise a TypeError
circle = Circle(5)
rectangle = Rectangle(4, 6)

print(f"Area of Circle: {circle.area()}")
print(f"Area of Rectangle: {rectangle.area()}")
```

```python
#3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car and further derive a class ElectricCar that adds a battery attribute.

class Vehicle:
    def __init__(self, type):
        self.type = type

class Car(Vehicle):
    def __init__(self, type, model):
        super().__init__(type)
        self.model = model

class ElectricCar(Car):
    def __init__(self, type, model, battery_size):
        super().__init__(type, model)
        self.battery_size = battery_size

# Example usage
electric_car = ElectricCar("Sedan", "Tesla Model S", "100 kWh")
print(f"Vehicle Type: {electric_car.type}")
print(f"Car Model: {electric_car.model}")
print(f"Battery Size: {electric_car.battery_size}")
```

```python
#4. Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes Sparrow and Penguin that override the fly() method.

class Bird:
    def fly(self):
        print("Most birds can fly.")

class Sparrow(Bird):
    def fly(self):
        print("Sparrows can fly.")

class Penguin(Bird):
    def fly(self):
        print("Penguins cannot fly.")

# Example usage
bird = Bird()
sparrow = Sparrow()
penguin = Penguin()

bird.fly()
sparrow.fly()
penguin.fly()
```

```python
#5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.

class BankAccount:
    def __init__(self, balance=0):
        self.__balance = balance  # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: ${amount}. New balance: ${self.__balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew: ${amount}. New balance: ${self.__balance}")
        elif amount > self.__balance:
            print("Insufficient funds.")
        else:
            print("Withdrawal amount must be positive.")

    def get_balance(self):
        return self.__balance

# Example usage
account = BankAccount(1000)
account.deposit(500)
account.withdraw(200)
account.withdraw(2000) # Insufficient funds
print(f"Current balance: ${account.get_balance()}")

# Trying to access the private attribute directly (will cause an AttributeError)
# print(account.__balance)
```

```python
#6. Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar and Piano that implement their own version of play().

class Instrument:
    def play(self):
        print("Playing an instrument.")

class Guitar(Instrument):
    def play(self):
        print("Playing the guitar.")

class Piano(Instrument):
    def play(self):
        print("Playing the piano.")

# Example usage
def make_instrument_play(instrument):
    instrument.play()

instrument = Instrument()
guitar = Guitar()
piano = Piano()

make_instrument_play(instrument)
make_instrument_play(guitar)
make_instrument_play(piano)
```

```python
#7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static method subtract_numbers() to subtract two numbers.

class MathOperations:
    @classmethod
    def add_numbers(cls, x, y):
        return x + y

    @staticmethod
    def subtract_numbers(x, y):
        return x - y

# Example usage
print(f"Addition: {MathOperations.add_numbers(10, 5)}")
print(f"Subtraction: {MathOperations.subtract_numbers(10, 5)}")
```

```python
#8. Implement a class Person with a class method to count the total number of persons created.

class Person:
    count = 0  # Class variable to keep track of the number of instances

    def __init__(self, name):
        self.name = name
        Person.count += 1  # Increment the count when a new object is created

    @classmethod
    def total_persons(cls):
        return cls.count

# Example usage
person1 = Person("Alice")
person2 = Person("Bob")
person3 = Person("Charlie")

print(f"Total number of persons created: {Person.total_persons()}")
```

```python
#9. Write a class Fraction with attributes numerator and denominator. Override the str method to display the fraction as "numerator/denominator".

class Fraction:
    def __init__(self, numerator, denominator):
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"

# Example usage
fraction = Fraction(3, 4)
print(fraction)
```

```python
#10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors.

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        if isinstance(other, Vector):
            return Vector(self.x + other.x, self.y + other.y)
        else:
            raise TypeError("Can only add a Vector to another Vector")

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

# Example usage
vector1 = Vector(2, 3)
vector2 = Vector(4, 5)
vector3 = vector1 + vector2
print(vector3)
```

```python
#12. Implement a class Student with attributes name and grades. Create a method average_grade() to compute the average of the grades.

class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        if not self.grades:
            return 0  # Return 0 if there are no grades
        return sum(self.grades) / len(self.grades)

# Example usage
student1 = Student("Alice", [85, 90, 78, 92])
student2 = Student("Bob", [70, 75, 80])

print(f"{student1.name}'s average grade: {student1.average_grade()}")
print(f"{student2.name}'s average grade: {student2.average_grade()}")
```

```python
#14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.

class Employee:
    def __init__(self, hours_worked, hourly_rate):
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate

class Manager(Employee):
    def __init__(self, hours_worked, hourly_rate, bonus):
        super().__init__(hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        base_salary = super().calculate_salary()
        return base_salary + self.bonus

# Example usage
employee = Employee(40, 20)
manager = Manager(40, 20, 500)

print(f"Employee Salary: ${employee.calculate_salary()}")
print(f"Manager Salary: ${manager.calculate_salary()}")
```

```python
#15. Create a class Product with attributes name, price, and quantity. Implement a method total_price() that calculates the total price of the product.

class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def total_price(self):
        return self.price * self.quantity

# Example usage
product = Product("Laptop", 1000, 2)
print(f"Total price of {product.name}: ${product.total_price()}")
```

```python
#16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        return "Moo"

class Sheep(Animal):
    def sound(self):
        return "Baa"

# Example usage
# animal = Animal() # This would raise a TypeError
cow = Cow()
sheep = Sheep()

print(f"Cow says: {cow.sound()}")
print(f"Sheep says: {sheep.sound()}")
```

```python
#17. Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that returns a formatted string with the book's details.

class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        return f"Title: {self.title}, Author: {self.author}, Year Published: {self.year_published}"

# Example usage
book = Book("The Hitchhiker's Guide to the Galaxy", "Douglas Adams", 1979)
print(book.get_book_info())
```

```python
#18. Create a class House with attributes address and price. Create a derived class Mansion that adds an attribute number_of_rooms.

class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

# Example usage
mansion = Mansion("123 Luxury Lane", 5000000, 20)
print(f"Mansion Address: {mansion.address}")
print(f"Mansion Price: ${mansion.price}")
print(f"Number of Rooms: {mansion.number_of_rooms}")
```
